```python
# Read the Delta table as a streaming source
rcb_stream = (
    spark.readStream
         .format("delta")
         .table("ashuu.rcb.rcb_2025_squad")
)

# Example transformation (optional)
rcb_filtered = rcb_stream.filter(rcb_stream.Age > 25)

# Write the stream into a Volume path with a supported trigger
query = (
    rcb_filtered.writeStream
                .format("delta")
                .outputMode("append")
                .option("checkpointLocation", "/Volumes/workspace/default/abcd/checkpoints")
                .option("path", "/Volumes/workspace/default/abcd")
                .trigger(availableNow=True)  # ðŸ‘ˆ supported trigger for your cluster
                .start()
)
```
